```python
import pandas as pd
```

```python
file = pd.read_csv('Employees.csv')
file
```

```python
file_cleaned = file.drop_duplicates()
file_cleaned
```

```python
file_cleaned['Age'] = file_cleaned['Age'].astype(int)
file_cleaned
```

```python
file_cleaned.rename(columns={'Salary(USD)': 'Salary(EGP)'}, inplace = True)
file_cleaned['Salary(EGP)'] = file_cleaned['Salary(EGP)'] * 50
file_cleaned
```

```python
average_age = file_cleaned['Age'].mean()

median_salary_egp = file_cleaned['Salary(EGP)'].median()

gender_counts = file_cleaned['Gender'].value_counts()
male_count = gender_counts.get('M', 0)
female_count = gender_counts.get('F', 0)
if female_count > 0:
    male_female_ratio = male_count / female_count
else:
    male_female_ratio = "undefined (no females in dataset)"

# Print the stats
print(f"Average Age: {average_age:.2f} years")
print(f"Median Salary (EGP): {median_salary_egp:.2f}")
print(f"Number of Male employees: {male_count}")
print(f"Number of Female employees: {female_count}")
print(f"Male to Female Ratio: {male_female_ratio}")
```

```python
file_cleaned.to_csv('Employees_New.csv', index=False)
```
